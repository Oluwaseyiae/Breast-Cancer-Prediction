### Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

### Import the Dataset

```python
dataset = pd.read_csv("dataset/data.csv")
dataset
```

### Separate the dependent and independent variables

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### Split dataset into train set and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

### Training the Support Vector Machine Model on the train set

```python
from sklearn.svm import SVC

classifier = SVC(kernel = "linear",random_state = 0)
classifier.fit(X_train_scaled, y_train)
```

### Predicting the result of the train set

```python
y_pred = classifier.predict(X_test_scaled)
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Checking the ratio of right and wrong 
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Accuray 
accuracy_score(y_test, y_pred)
```

### Comparing y_pred and y_train

```python
y_pred = [f"{val:.0f}" for val in y_pred]
comparison_df = pd.DataFrame({
    "Original Pice": y_test,
    "Predicted Price": y_pred
})

comparison_df.head(50)
```
